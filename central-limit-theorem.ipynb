```
# For two of the following distributions:

# Normal
# Binomial
# Gamma
# Poisson

# Create a line chart plotting the sample mean as more and more trials are 
# conducted, up to 1,000 trials.

# Randomly sample the mean 1,000 times and plot the resulting frequency of 
# sample means as a histogram.

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean

# Normal
norm_res = []
for num_trials in range(1,1000):
    trials = np.random.normal(0, 1, num_trials)
    norm_res.append(trials.mean())

norm_means = pd.DataFrame(norm_res)
norm_means.plot()
plt.show()

norm_samples = []
for _ in range(1000):
    norm_samples.append(mean(np.random.normal(0, 1, 1000)))

plt.hist(norm_samples, bins=10)
plt.show()

# Binomial
bino_res = []
for num_trials in range(1,1000):
    trials = np.random.binomial(10, 0.5, num_trials)
    bino_res.append(trials.mean())

bino_means = pd.DataFrame(bino_res)
bino_means.plot()
plt.show()

bino_samples = []
for _ in range(1000):
    bino_samples.append(mean(np.random.binomial(10, 0.5, 1000)))

plt.hist(bino_samples, bins=10)
plt.show()
```
