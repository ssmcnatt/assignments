```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

norm = np.random.normal(0, 1, 10000)
plt.hist(norm, bins=100)
plt.axvline(norm.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(norm.mean() + norm.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm.mean() - norm.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

bino = np.random.binomial(1, .5, 10000)
plt.hist(bino, bins=100)
plt.axvline(bino.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(bino.mean() + bino.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(bino.mean() - bino.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

logi = np.random.logistic(10, 1, 10000)
plt.hist(logi, bins=100)
plt.axvline(logi.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(logi.mean() + logi.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logi.mean() - logi.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

tria = np.random.triangular(-3, 0, 8, 10000)
plt.hist(tria, bins=100)
plt.axvline(tria.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(tria.mean() + tria.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(tria.mean() - tria.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

geom = np.random.geometric(0.33, 10000)
plt.hist(geom, bins=100)
plt.axvline(geom.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(geom.mean() + geom.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(geom.mean() - geom.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

gumb = np.random.gumbel(1, 0.1, 10000)
plt.hist(gumb, bins=100)
plt.axvline(gumb.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(gumb.mean() + gumb.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gumb.mean() - gumb.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()

```
