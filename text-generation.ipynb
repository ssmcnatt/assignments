## Train a Markov chain model by using only the negative sentiment tweets. Generate some random sentences. Do the generated sentences exhibit the same negative sentiment?


```python
#!pip install markovify
```

```python
import markovify
```

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
```

```python
all_doc = doc
```

```python
# Get negative and positive sentences
negative_doc = ''
positive_doc = ''
blob = TextBlob(doc)
for sentence in blob.sentences:
    if sentence.sentiment.polarity < 0:
        negative_doc = negative_doc + ' ' + sentence.raw
    if sentence.sentiment.polarity > 0.7:
        positive_doc = positive_doc + ' ' + sentence.raw
```

```python
# Parse the doc. This can take some time.
nlp = spacy.load('en')
nlp.max_length = 1500000
nlp_doc = nlp(negative_doc)
```

```python
# Group into sentences
# Use the sentences that have more than one character
doc_sents = " ".join([sent.text for sent in nlp_doc.sents if len(sent.text) > 1])
```

```python
doc_generator = markovify.Text(doc_sents, state_size = 3)
```

```python
# randomly generated sentences
for i in range(5):
    print(doc_generator.make_sentence(tries=100))

print('\n')
# randomly generated sentences of no more than 100 characters
for i in range(5):
    print(doc_generator.make_short_sentence(100))
```
